Estrutura if-else

```python
# Programa para verificar se um número é par ou impar
n = int(input())

if n % 2 == 0:
    r = 'é par'
else:
    r = 'é impar'

print(n, r)
```

Forma alternativa

```python
# Esta mesma expressão pode também ser expressa como
n = int(input())

r = 'é par' if n % 2 == 0 else 'é impar'

print(n, r)
```
